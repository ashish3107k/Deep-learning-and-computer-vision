```python
def visulaize(orignal,augmented):
  fig = plt.figure()
  plt.subplot(1,2,1)
  plt.title('orignal')
  plt.imshow(image)
  plt.subplot(1,2,2)
  plt.title('augmented')
  plt.imshow(augmented)

  
```

```python
flipped = tf.image.flip_left_right(image)
```

```python
plt.imshow(flipped)
```

```python
visulaize(image,flipped)
```

```python
grayscaled=tf.image.rgb_to_grayscale(image)
visulaize(image,tf.squeeze(grayscaled))
plt.colorbar()
```

```python
plt.imshow(tf.squeeze(grayscaled),cmap='gray_r')
```

```python
saturated = tf.image.adjust_saturation(image,5)
visulaize(image,saturated)
```

```python
bright = tf.image.adjust_brightness(image,0.4)
visulaize(image,bright)
```

```python
centercropped = tf.image.central_crop(image,central_fraction=0.5)
visulaize(image,centercropped)
```

```python
rotated = tf.image.rot90(image)
visulaize(image,rotated)
```

```python
contrass = tf.image.adjust_contrast(image,0.9)
visulaize(image,contrass)
```

```python
aug1 = tf.image.flip_up_down(image)
visulaize(image,aug1)
```
